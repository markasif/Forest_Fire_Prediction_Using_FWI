```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("Algerian_forest_fires_dataset_UPDATE.csv",header=1)
```

```python
df.head()
```

## Data Cleaning

```python
#missing values
df.isnull().sum()
```

```python
#missing values in any columns
df[df.isnull().any(axis=1)]
```

```python
## when i observed the data. the date contain two set based on religion from 122th index, we can make new columns based on the region
```

```python
df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
```

```python
df.info()
```

```python
df[["Region"]] = df[["Region"]].astype(int)
```

```python
df.head()
```

```python
##remove missinfg values
df= df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df= df.drop(122).reset_index(drop=True)
```

```python
df.iloc[122:]
```

```python
df.columns
```

```python
##removing white space in columns
df.columns = df.columns.str.strip()
df.columns
```

## changing data types of columns

```python
df.info()
```

```python
df.head()
```

```python
df[["day","month","year","Temperature","RH","Ws"]] = df[["day","month","year","Temperature","RH","Ws"]].astype(int)
```

```python
# Identify object-type columns
objects = [features for features in df.columns if df[features].dtypes == 'O']

# Convert columns to float (if not "Classes")
for i in objects:
    if i != "Classes":
        df[i] = df[i].astype(float)  # Assign the converted column back to the DataFrame
```

```python
df.info()
```

```python
df.to_csv('Algerian_forest_fire_cleaned.csv',index=False)
```
